```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
```

```python
#import curated tables exported from LIPIDMAPS (Description in methods section: Searches in LIPID MAPS)

head = pd.read_csv('/Users/helenarusso/Documents/Pesquisa/UCSD Post-doc/Acyl lipids/Manuscript/Supplementary Material/YASIN_LIPIDMAPS/Source_data_head.txt', sep='\t')
head['Head'] = head['Head'].str.capitalize()

tail = pd.read_csv('/Users/helenarusso/Documents/Pesquisa/UCSD Post-doc/Acyl lipids/Manuscript/Supplementary Material/YASIN_LIPIDMAPS/Source_data_tail.txt', sep='\t')
tail = tail.sort_values(by='order', ascending=True)
```

```python
# Create bar plot
plt.figure(figsize=(16, 6))
sns.barplot(x='Head', y='count', data=head, color='#3167B0', edgecolor='black')

# Add title and labels
plt.xlabel('Head group')
plt.ylabel('Counts')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Remove the top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('/.../lipidmaps_head.pdf')

plt.show()
```

```python
# Create bar plot
plt.figure(figsize=(12, 4))
sns.barplot(x='Carbons', y='count', data=tail, color='#BF2C23', edgecolor='black')

# Add title and labels
plt.xlabel('Number of carbons')
plt.ylabel('Counts')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Remove the top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('/.../lipidmaps_tail.pdf')

plt.show()
```
